```python
#importing dependencies
import pandas as pd
from pathlib import Path
import numpy as np
```

```python
#Setup
school_csv = Path("Resources/schools_complete.csv")
student_csv = Path("Resources/students_complete.csv")

#read the CSV files 
school_df = pd.read_csv(school_csv)
student_df = pd.read_csv(student_csv)

student_df
```

```python
school_df
```

```python
#Merge the two dataframes into 1 dataframe 
combined_df = pd.merge(student_df, school_df, on="school_name", how='outer')
combined_df
```

## District Summary

```python
#Calculate total number of unique schools
unique_schools = combined_df["school_name"].unique()
total_schools = len(unique_schools)

total_schools
```

```python
#Calculate the total number of students

total_students = combined_df["student_name"].count()

total_students
```

```python
#Calculate the total budget
#Need to look at the budget column in the school dataframe since you know what the budget allocation is per school

total_budget = school_df["budget"].sum()
total_budget
```

```python
#Calculate the average math score
average_math_score = combined_df["math_score"].mean()
average_math_score
```

```python
#Calculate the average reading score
average_reading_score = combined_df["reading_score"].mean()
average_reading_score
```

```python
students_passed_math = combined_df[combined_df["math_score"] >= 70]
students_passed_math
```

```python
number_students_passed_math = students_passed_math["math_score"].count()
number_students_passed_math
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
students_passed = combined_df[combined_df["math_score"] >= 70]
number_students_passed = students_passed["math_score"].count()
percentage_passed_math = (number_students_passed/total_students) * 100
percentage_passed_math
```

```python
students_passed_reading = combined_df[combined_df["reading_score"] >= 70]
students_passed_reading
```

```python
number_students_passed_reading = students_passed_reading["reading_score"].count()
number_students_passed_reading
```

```python
# Use the following to calculate the percentage of students who passed reading (reading scores greather than or equal to 70)
students_passed_reading = combined_df[combined_df["reading_score"] >= 70]
number_students_passed_reading = students_passed_reading["reading_score"].count()
percentage_passed_reading = (number_students_passed_reading/total_students) * 100
percentage_passed_reading
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = combined_df[(combined_df["math_score"] >= 70) & (combined_df["reading_score"] >= 70)].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(total_students) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame({"Total Schools": [total_schools], "Total Students": [total_students], 
                                "Total Budget": [total_budget], "Average Math Score": [average_math_score],
                                "Average Reading Score": [average_reading_score], "% Passing Math": [percentage_passed_math],
                                "% Passing Reading": [percentage_passed_reading], "Overall Passing Rate": [overall_passing_rate]})

# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
district_summary
```

## School Summary

```python
#Use the code provided to select all of the school types
school_types = school_df.loc[:,["school_name","type"]]

school_types
```

```python
#Calculate the total student count per school

per_school_counts = school_df.loc[:,["school_name","size"]]

total_students_per_school = per_school_counts.rename(columns={"size": "Total Students"})

total_students_per_school
```

```python
#Calculate the total budget per school

per_school_budget = school_df.loc[:,["school_name","size","budget"]]

total_budget = per_school_budget.rename(columns = {"size":"Total Students","budget":"Total Budget"})

total_budget = total_budget.sort_values("school_name")

total_budget
```

```python
# Calculate the total per capita spending per school

total_budget["Per Student Budget"] = (total_budget["Total Budget"] / total_budget["Total Students"])

total_budget
```

```python
#First we need to create a dataframe that has just the school name, reading score and math score by schools

math_reading_score_group = combined_df.loc[:,["school_name","reading_score","math_score"]]

math_reading_score_group
```

```python
# Calculate the average test scores per school
per_school = math_reading_score_group.groupby(["school_name"]).mean()

average_per_school = per_school.rename(columns = {"reading_score": "Average Reading Score","math_score": "Average Math Score"})

average_per_school
```

```python
# Calculate the number of students per school with math scores of 70 or higher

math_group = combined_df.loc[:,["school_name","math_score"]]

students_passing_math = math_group[math_group["math_score"] >= 70]

school_students_passing_math = students_passing_math.groupby(["school_name"]).count()

school_students_passing_math = school_students_passing_math.rename(columns = {"math_score": "Total Students who Passed Math"})

school_students_passing_math
```

```python
# Calculate the number of students per school with reading scores of 70 or higher

reading_group = combined_df.loc[:,["school_name","reading_score"]]

students_passing_reading = reading_group[reading_group["reading_score"] >= 70]

school_students_passing_reading = students_passing_reading.groupby(["school_name"]).count()

school_students_passing_reading = school_students_passing_reading.rename(columns = {"reading_score": "Total Students who Passed Reading"})

school_students_passing_reading
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher

students_passing_math_and_reading = math_reading_score_group[(math_reading_score_group["reading_score"] >= 70) & (math_reading_score_group["math_score"] >= 70)]


school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).size()

school_students_passing_math_and_reading
```

```python
# DataFrame to show number of students who passed both math and reading
test_df = math_reading_score_group[(math_reading_score_group["reading_score"] >= 70) & (math_reading_score_group["math_score"] >= 70)]

#By using count, this returns a format in a dataframe that I can call
passing_math_and_reading_df = test_df.groupby(["school_name"]).count()

passing_math_and_reading_df = passing_math_and_reading_df.rename(columns = {"reading_score":"# of Students who passed Reading and Math"})

del passing_math_and_reading_df["math_score"]

passing_math_and_reading_df
```

```python
# Merging the dataframes that contain the total students who passed math and the total students who passed reading per school

new_df = pd.merge(school_students_passing_math, school_students_passing_reading, on="school_name", how='outer')

new2_df = pd.merge(total_students_per_school, new_df, on="school_name", how='outer')

new3_df = pd.merge(passing_math_and_reading_df, new2_df, on="school_name", how='outer')

new3_df = new3_df.sort_values("school_name")

new3_df
```

```python
# Use the provided code to calculate the passing rates per school
new3_df["% Passing Math"] = (new3_df["Total Students who Passed Math"] / new3_df["Total Students"]) * 100

new3_df["% Passing Reading"] = (new3_df["Total Students who Passed Reading"] / new3_df["Total Students"]) * 100


new3_df["% Overall Passing"] = (new3_df["# of Students who passed Reading and Math"] / new3_df["Total Students"]) * 100
```

```python
new3_df
```

```python
# Merging the different dataframes that I created to incorporate the final columns 

merge1_df = pd.merge(school_types,total_budget, on="school_name", how = "outer")

merge2_df = pd.merge(merge1_df, average_per_school, on="school_name", how = "outer")

merge3_df = pd.merge(merge2_df, new3_df, on="school_name", how = "outer")

del merge3_df["# of Students who passed Reading and Math"]
del merge3_df["Total Students_y"]
del merge3_df["Total Students who Passed Math"]
del merge3_df["Total Students who Passed Reading"]

merge3_df = merge3_df.rename(columns = {"type": "School Type","Total Students_x":"Total Students", "Total Budget":"Total School Budget"})

merge3_df = merge3_df.sort_values("school_name")
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
per_school_summary = merge3_df.set_index("school_name")
per_school_summary.index.name = None

# Formatting
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

# Display the DataFrame
per_school_summary
```

## Highest-Performing Schools (by % Overall Passing)

```python
# Sort the schools by `% Overall Passing` in descending order and display the top 5 rows.

top_schools = per_school_summary.sort_values("% Overall Passing", ascending=False)

top_schools.head(5)
```

## Bottom Performing Schools (By % Overall Passing)

```python
# Sort the schools by `% Overall Passing` in ascending order and display the top 5 rows.

bottom_schools = per_school_summary.sort_values("% Overall Passing", ascending=True)
bottom_schools.head(5)
```

## Math Scores by Grade

```python
# Use the code provided to separate the data by grade
ninth_graders = combined_df[(combined_df["grade"] == "9th")]
tenth_graders = combined_df[(combined_df["grade"] == "10th")]
eleventh_graders = combined_df[(combined_df["grade"] == "11th")]
twelfth_graders = combined_df[(combined_df["grade"] == "12th")]
```

```python
# Group by `school_name` and take the mean of the `math_score` column for each.
ninth_graders_math = ninth_graders.loc[:,["school_name","math_score"]]
ninth_grade_math_scores = ninth_graders_math.groupby("school_name").mean()
ninth_grade_math_scores = ninth_grade_math_scores.rename(columns={"math_score":"9th"})

tenth_graders_math = tenth_graders.loc[:,["school_name","math_score"]]
tenth_grade_math_scores = tenth_graders_math.groupby("school_name").mean()
tenth_grade_math_scores = tenth_grade_math_scores.rename(columns={"math_score":"10th"})

eleventh_graders_math = eleventh_graders.loc[:,["school_name","math_score"]]
eleventh_grade_math_scores = eleventh_graders_math.groupby("school_name").mean()
eleventh_grade_math_scores = eleventh_grade_math_scores.rename(columns={"math_score":"11th"})

twelfth_graders_math = twelfth_graders.loc[:,["school_name","math_score"]]
twelfth_grade_math_scores = twelfth_graders_math.groupby("school_name").mean()
twelfth_grade_math_scores = twelfth_grade_math_scores.rename(columns={"math_score":"12th"})
```

```python
# Combine each of the scores above into single DataFrame called `math_scores_by_grade`

math1 = pd.merge(ninth_grade_math_scores, tenth_grade_math_scores, on="school_name", how="outer")

math2 = pd.merge(math1, eleventh_grade_math_scores, on="school_name", how="outer")

math_scores_by_grade = pd.merge(math2, twelfth_grade_math_scores, on="school_name", how="outer" )
```

## Scores by School Spending

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]

spending_labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the merge3_df since the values don't have dollar signs in the "Per Student Budget" column
school_spending_df = merge3_df.copy()
school_spending_df
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Per Student Budget"], 
                                              spending_bins, labels=spending_labels, 
                                              include_lowest=True)

school_spending_df
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()
```

```python
# Assemble into DataFrame
spending_dict = {'Average Math Score': spending_math_scores,
                'Average Reading Score': spending_reading_scores,
                '% Passing Math': spending_passing_math,
                '% Passing Reading': spending_passing_reading,
                '% Overall Passing': overall_passing_spending}
spending_summary = pd.DataFrame(spending_dict)

# Display results
spending_summary
```

## Scores by School Size

```python
# Calculate averages for the desired columns. 
size_math_scores = per_school_summary.groupby(["School Size"])["Average Math Score"].mean()
size_reading_scores = per_school_summary.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = per_school_summary.groupby(["School Size"])["% Passing Math"].mean()
size_passing_reading = per_school_summary.groupby(["School Size"])["% Passing Reading"].mean()
size_overall_passing = per_school_summary.groupby(["School Size"])["% Overall Passing"].mean()
```

```python
# Assemble into DataFrame
size_dict = {'Average Math Score': size_math_scores,
                'Average Reading Score': size_reading_scores,
                '% Passing Math': size_passing_math,
                '% Passing Reading': size_passing_reading,
                '% Overall Passing': size_overall_passing}
size_summary = pd.DataFrame(size_dict)

# Display results
size_summary
```
